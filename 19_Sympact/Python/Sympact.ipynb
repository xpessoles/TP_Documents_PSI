# Etude de la barrière Sympact

### Paramètres géométriques 


```python
import numpy as np
import matplotlib.pyplot as plt
R = 0.081 #m
H = 0.112 #m
alpha = np.radians(45)

theta_0 = -np.radians(30.61)
theta_f = -np.radians(30.61)+np.radians(246)
```

## Loi Entrée-Sortie géométrique
On note $\theta$ l'angle du moteur et $\varphi$ l'angle de la barrière.

On montre que $\tan\varphi(t)=\dfrac{H+R\sin\theta(t)}{R\cos\theta(t)}$.

```python
les_theta = np.linspace(theta_0,theta_f,1000)
les_phi = np.arctan2(H+R*np.sin(les_theta),R*np.cos(les_theta))
plt.grid()
plt.xlabel("$\\theta$ Angle moteur" )
plt.ylabel("$\\varphi$ Angle barriere" )
plt.plot(np.degrees(les_theta),np.degrees(les_phi))
```
