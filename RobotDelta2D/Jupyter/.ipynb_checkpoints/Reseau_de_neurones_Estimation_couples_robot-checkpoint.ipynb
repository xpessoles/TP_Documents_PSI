# Simulation du robot

La première section est une simulation mécanique d'un robot série à 2 DDL
bras_robot2D.png

Les deux sections suivantes simulent la cinématique et la dynamique du robot ci-dessus.
Nous l'utiliserons comme une "boite noire" afin de générer des données liées à la dynamique du robot

```python
# partie simulation mécanique inspiré par "Etude et contrôle d’un bras de robot 2D Modélisation sous python avec sympy mechanics" Marc BUFFAT, dpt mécanique, Université Claude Bernard Lyon 1
# https://perso.univ-lyon1.fr/marc.buffat/2019/Robotique_Bras2D/index.html

#Section paramétrage du modèle

%matplotlib inline
%autosave 300
import numpy as np
import scipy as sp
import sympy as sp
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 14
from IPython.core.display import HTML
from IPython.display import display,Image
from matplotlib import animation

from sympy.physics.vector import init_vprinting
init_vprinting(use_latex='mathjax', pretty_print=False)

# Nombre de pas de temps
Np = 50
# longueur du bras en m
L1 = 0.5
L2 = 0.5

#Les modèles cinématiques et dynamiques sont résolus analytiquement

#définition des paramêtres du problème et des DDL
#définition des points et des repéres
#définition de la position (dans les bons repères)

from sympy.physics.mechanics import dynamicsymbols, Point, ReferenceFrame

l1, l2 = sp.symbols('l1 l2')
theta1, theta2 = dynamicsymbols('theta1 theta2')

RO = ReferenceFrame('R_O')
RP = ReferenceFrame('R_P')
RQ = ReferenceFrame('R_Q')

RP.orient(RO, 'Axis', [theta1, RO.z])
RQ.orient(RP, 'Axis', [theta2, RO.z])

T02 = RO.dcm(RQ)
T02.simplify()

O = Point('O')
P = Point('P')
Q = Point('Q')

P.set_pos(O, l1 * RP.x)
Q.set_pos(P, l2 * RQ.x)

Q.pos_from(O)

qxy = (Q.pos_from(O).express(RO)).simplify()

qx=qxy.dot(RO.x)
qy=qxy.dot(RO.y)

pxy = (P.pos_from(O).express(RO)).simplify()

px = pxy.dot(RO.x)
py = pxy.dot(RO.y)

#Simulation géométrique
#conversion des formules analytiques en fonction python (numérique) avec lambdify

#fonction Qx,Qy calculant les coordonnées de Q dans RO
#fonction Px,Py calculant les coordonnées de P dans RO
#les paramêtres sont les symbols intervenant dans les formules

Qx = sp.lambdify((l1, l2, theta1, theta2), qx, 'numpy')
Qy = sp.lambdify((l1, l2, theta1, theta2), qy, 'numpy')
Px = sp.lambdify((l1, theta1), px, 'numpy')
Py = sp.lambdify((l1, theta1), py, 'numpy')


#Fonction pour l'animation de la trajectoire
def animate(i):
    global PX,PY,QX,QY
    line1.set_data([0.,PX[i]],[0.,PY[i]])
    line2.set_data([PX[i],QX[i]],[PY[i],QY[i]])
    pt1.set_offsets([QX[i]+0.05,QY[i]])
    return line1,line2,pt1
```

### Section contenant la génération de commande et la résolution dynamique


```python

# symbols et inertie
from sympy.physics.mechanics import inertia
I1x,I1y = sp.symbols('I_1x I_1y')
I1 = inertia(RP,I1x,I1y,I1y)

C1,C2,m,m1,m2,g = sp.symbols('C1 C2 m m_1 m_2 g')
# centre de gravité
G1 = Point('G1')
G1.set_pos(O,l1/2 * RP.x)
yg1=G1.pos_from(O).express(RO).dot(RO.y)
G2 = Point('G2')
G2.set_pos(P,l2/2 * RQ.x)
yg2=G2.pos_from(O).express(RO).dot(RO.y)
O.set_vel(RO,0.)
# vitesse du centre de gravité G1
G1.v2pt_theory(O,RO,RP)
G1.pos_from(O), G1.vel(RO)

#Calcul de la quantité de mouvement, du moment cinétique
from sympy.physics.mechanics import RigidBody
TigeOP = RigidBody('Tige_OP',G1,RP,m1,(I1,G1))
from sympy.physics.mechanics import linear_momentum, angular_momentum
linear_momentum(RO,TigeOP), angular_momentum(G1,RO,TigeOP)

#Energie cinétique et potentiel
from sympy.physics.mechanics import kinetic_energy, potential_energy
kinetic_energy(RO, TigeOP)

TigeOP.potential_energy = m1*g*G1.pos_from(O).dot(RO.y)
potential_energy(TigeOP)

m1*G1.acc(RO)

I2x,I2y,m = sp.symbols('I_2x I_2y m')
I2 = inertia(RQ,I2x,I2y,I2y)
# vitesse du centre de gravité
G2.v2pt_theory(P,RO,RQ)

TigePQ = RigidBody('Tige_PQ',G2,RQ,m2,(I2,G2))
linear_momentum(RO,TigePQ), angular_momentum(G2,RO,TigePQ)

kinetic_energy(RO,TigePQ)

# on ajoute la masse M en Q
TigePQ.potential_energy = m2*g*G2.pos_from(O).dot(RO.y) + m*g*Q.pos_from(O).dot(RO.y)
potential_energy(TigePQ)

kinetic_energy(RO,TigeOP,TigePQ).simplify()

potential_energy(TigeOP,TigePQ).simplify()

#Calcul du Lagrangien
from sympy.physics.mechanics import Lagrangian, LagrangesMethod
L = Lagrangian(RO,TigeOP,TigePQ).simplify()

LM = LagrangesMethod(L,[theta1,theta2], forcelist = [(RP, (C1-C2)*RO.z),(RQ, C2*RO.z)], frame = RO)
eqs = LM.form_lagranges_equations()

C=sp.Array([[C1],[C2]])
MQ=LM.mass_matrix
from sympy.matrices import Matrix
Csympy=Matrix(C)
FQ=LM.forcing-Csympy

#génération d'une commande par retour d'état

q = sp.Function('q')
t = sp.Symbol('t',positive=True)
kv = sp.symbols('k_v',positive=True)
qd,q0 = sp.symbols('q_d q_0')
eq = sp.Eq(sp.Derivative(q(t),t,t) + kv*sp.Derivative(q(t),t) + kv*kv/4*(q(t)-qd) , 0 )

solq=sp.dsolve(eq, q(t), ics={q(0):q0, q(t).diff(t).subs(t,0):0})
qs=sp.Lambda(t,solq.rhs)

theta1d,theta10,theta2d,theta20 = sp.symbols('theta_1_d theta_1_0 theta_2_d theta_2_0')
theta1s=sp.Lambda(t,qs(t).subs([(qd,theta1d),(q0,theta10)]))
theta2s=sp.Lambda(t,qs(t).subs([(qd,theta2d),(q0,theta20)]))

u1 = -kv*sp.diff(theta1s(t),t) + kv*kv/4*(theta1s(t)-theta1d)
u1s = sp.Lambda(t,u1.simplify())
u2 = -kv*sp.diff(theta2s(t),t) + kv*kv/4*(theta2s(t)-theta2d)
u2s = sp.Lambda(t,u2.simplify())

params = [(g,9.81), (m,10), (m1,1), (m2,1), (l1, L1), (l2, L2)]
# liste supplémentaire des parametres
params=[ (I1y, m1*(0.02**2/4. + l1**2/12.)), (I2y, m2*(0.02**2/4. + l2**2/12.))] + params

# notation pour les dérivées premières
theta1p, theta2p = dynamicsymbols('theta1 theta2',level=1)
# fonctions python
Theta1s = sp.lambdify((t,theta10,theta1d,kv),theta1s(t))
Theta2s = sp.lambdify((t,theta20,theta2d,kv),theta2s(t))
dTheta1s = sp.lambdify((t,theta10,theta1d,kv),sp.diff(theta1s(t),t))
dTheta2s = sp.lambdify((t,theta20,theta2d,kv),sp.diff(theta2s(t),t))
U1s     = sp.lambdify((t,theta10,theta1d,kv),theta1s(t))
U2s     = sp.lambdify((t,theta20,theta2d,kv),theta2s(t))
MQs     = sp.lambdify((theta1,theta2,theta1p,theta2p),MQ.subs(params))
FQs     = sp.lambdify((theta1,theta2,theta1p,theta2p),FQ.subs(params))

Theta10 = np.pi
Theta20 = -np.pi*2
Kv = 10.0

def Thetas(t, Theta10, Theta20, Theta1p, Theta2p):
    """solution a t"""
    return np.array([Theta1s(t,Theta10,Theta1p,Kv),Theta2s(t,Theta20,Theta2p,Kv)])
def dThetas(t, Theta10, Theta20, Theta1p, Theta2p):
    """dérivée de la solution a t"""
    return np.array([dTheta1s(t,Theta10,Theta1p,Kv),dTheta2s(t,Theta20,Theta2p,Kv)])

#Fonction qui retourne les couples articulaires
def Couples(T, Theta10, Theta20, Theta1p, Theta2p):
    C = np.zeros((2,T.size))
    for k,t in enumerate(T):
        theta1  = Theta1s(t,Theta10,Theta1p,Kv)
        dtheta1 = dTheta1s(t,Theta10,Theta1p,Kv)
        theta2  = Theta2s(t,Theta20,Theta2p,Kv)
        dtheta2 = dTheta2s(t,Theta20,Theta2p,Kv)
        U = np.array([U1s(t,Theta10,Theta1p,Kv),U2s(t,Theta20,Theta2p,Kv)])
        M = MQs(theta1,theta2,dtheta1,dtheta2)
        F = FQs(theta1,theta2,dtheta1,dtheta2)[:,0]
        C[:,k] = np.dot(M,U)-F
    return C

def cinematique_inv(x2,y2):
    '''calcul les angles theta1,theta2 du bras en fct de position fixée x2,y2'''
    global L1,L2
    L = np.sqrt(x2**2+y2**2)
    thetap = np.arccos(x2/L)
    theta1 = np.arccos((L1**2 + L**2  - L2**2)/(2*L1*L)) + thetap
    theta2 = np.arccos((L1**2 + L2**2 - L**2) /(2*L1*L2)) - np.pi
    return theta1,theta2

#Choix d'une position finale X,Y
X, Y = (L1+L2)/3,L2/3
Theta1p,Theta2p = cinematique_inv(X,Y)

#Calcul du modèle pour avoir les angles, les couples et les positions à chaque pas de temps
T = np.linspace(0,2.0,Np)
THETA = Thetas(T, Theta10, Theta20, Theta1p, Theta2p)
CC    = Couples(T, Theta10, Theta20, Theta1p, Theta2p)
X1 = Px(L1,THETA[0,:])
Y1 = Py(L1,THETA[0,:])
X2 = Qx(L1,L2,THETA[0,:],THETA[1,:])
Y2 = Qy(L1,L2,THETA[0,:],THETA[1,:])

#Affichage de la trajectoire et des couples articulaires
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.plot(X1,Y1,label='P')
plt.plot(X2,Y2,label='Q')
plt.xlabel("X (en m)")
plt.ylabel("Y (en m)")
plt.title('trajectoire')
plt.plot([0,X1[-1],X2[-1]],[0.,Y1[-1],Y2[-1]],'-ok',lw=10,markersize=25)
plt.legend()
plt.axis('equal')
plt.subplot(1,2,2)
plt.plot(T,CC[0,:],label='C1')
plt.plot(T,CC[1,:],label='C2')
plt.xlabel("Temps (en s)")
plt.ylabel("Couple (en Nm)")
plt.legend()
plt.title("Commandes du bras");


theta1s = THETA[0,:]
theta2s = THETA[1,:]
# positions fct angle
QX = np.array(Qx(L1, L2, theta1s, theta2s))
QY = np.array(Qy(L1, L2, theta1s, theta2s))
PX = np.array(Px(L1,theta1s))
PY = np.array(Py(L1,theta1s))

#Paramètres de la fenetre d'animation
Fig=plt.figure(figsize=(8,8))
ax = Fig.add_subplot(111, aspect='equal')
ax.set_axis_off()
ax.set_xlim((-1.2*(L1+L2),1.2*(L1+L2)))
ax.set_ylim((-1.2*(L1+L2),1.2*(L1+L2)))
ax.set_title('mouvement du bras de robot',fontsize=30)
Fig.set_facecolor("#ffffff")
line1, = ax.plot([0.,L1], [0.,0.], 'o-b', lw=18 , markersize=25)
line2, = ax.plot([L1,L1+L2], [0.,0.], 'o-', lw=18 , markersize=25)
pt1    = ax.scatter([L1+L2+0.05],[0.],marker="$\in$",s=800,c="black",zorder=3)

anim = animation.FuncAnimation(Fig, animate, np.arange(1, Np), interval=40, blit=True)
```

Condensé des fonctions précédente pour lancer en une fonction une simulation dynamique : 

C'est ce qui va nous permettre de le lancer n fois et avoir plein d'exemples pour notre base d'apprentissage

```python
def simulation_dynamique(Theta10,Theta20,X, Y): 

    Theta1p,Theta2p = cinematique_inv(X,Y)

    T = np.linspace(0,2.0,Np)
    THETA = Thetas(T, Theta10, Theta20, Theta1p, Theta2p)
    CC    = Couples(T, Theta10, Theta20, Theta1p, Theta2p)
    X1 = Px(L1,THETA[0,:])
    Y1 = Py(L1,THETA[0,:])
    X2 = Qx(L1,L2,THETA[0,:],THETA[1,:])
    Y2 = Qy(L1,L2,THETA[0,:],THETA[1,:])

    
    return CC, THETA, X1, Y1, X2, Y2, 
```

Exemple de simulation dynamique avec : $\theta_{1_0}$ = 0.4 ; $\theta_{2_0}$ = 0 ; $X$ = -0.2 ; $Y$ = -0.3; 


*   $\theta_{1_0}$ est l'angle $\theta_{1}$ initial
*   $\theta_{2_0}$ est l'angle $\theta_{2}$ initial
*   $(X,Y)$ sont les coordonnées de la position finale de l'effecteur (point Q)


```python
CC, THETA, X1, Y1, X2, Y2=simulation_dynamique(0.4,0,-0.2, -0.3)

#La fonction "anim" utilise les variables globales PX, PY, QX et QY. Il faut donc les mettre a jour
PX=X1
PY=Y1
QX=X2
QY=Y2
anim = animation.FuncAnimation(Fig, animate, np.arange(1, Np), interval=40, blit=True)
```

#Génération d'une base de données

Ici, nous utilisons la simulation pour générer des valeurs d'angles, de couples et de trajectoires.

Nous réalisons n itérations de la simulation avec des conditions initiales et finales aléatoires.

Les valeurs que nous metterons en entrée du réseau : $\theta_1$ et $\theta_2$

Les valeurs à estimer : $C_1$ et $C_2$ 

Ce sont des valeurs échantillonnées sur $N_p$ mesures par simulation. Donc pour chaque simulation, l'entrée est de dimension $(2,N_p)$ et la sorite $(2,N_p)$. Valeur par défaut de $N_p=50$ modifiable plus haut

Remarque :

On fait ca car nous n'avons pas de mesures sur un système réel. C'est un peu le point clé ici :  Il faut immaginer qu'on ne sais pas simuler le comportement du système mais que l'on peut réaliser des mesures dessus. 

Après la section suivante, nous ne faisons plus jamais appel à la simulation.

```python
n=10  # n=10 permet de visualiser du surapprentissage. phénomène absent à n=1000

#on génère n exemples de conditions initiales theta_0 et finales X,Y
cl=np.random.rand(4,n)
#Normalisation pour explorer tous l'espace de travail
cl=np.multiply(cl,np.array([[2*np.pi],[2*np.pi],[(L1+L2)*(np.sqrt(2)/2)],[(L1+L2)*(np.sqrt(2)/2)]]))

#rappel : Np est le nombre de pas de temps dans un mouvement

Ctot = np.zeros((n,2, Np))
Thetatot = np.zeros((n,2, Np))
X1tot = np.zeros((n,1, Np))
Y1tot = np.zeros((n,1, Np))
X2tot = np.zeros((n,1, Np))
Y2tot = np.zeros((n,1, Np))


for i in range(0,n-1):
    Ctot[i], Thetatot[i], X1tot[i], Y1tot[i], X2tot[i], Y2tot[i]=simulation_dynamique(cl[0,i],cl[1,i],cl[2,i], cl[3,i])

```

#Début du machine learning

```python
#Chargement des bibliothèques
from tensorflow.keras.layers import Input, Dense, Activation, Dropout, Flatten, LSTM
from tensorflow.keras.models import Model
import tensorflow as tf
```

Séparation des données en une base d'apprentissage et une base de validation

```python
X=Thetatot
y=Ctot

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Choix de la structure du réseau

```python
input_layer = Input(shape=(X.shape[1:3]))
dense_layer_1 = Dense(150, activation='relu')(input_layer)
dense_layer_2 = Dense(100, activation='relu')(dense_layer_1)
output = Dense(y.shape[2], activation='linear')(dense_layer_2)

model = Model(inputs=input_layer, outputs=output)
model.compile(loss='MSE', optimizer='adam', metrics=[tf.keras.metrics.MeanSquaredError()]) #MSE = mean square error -> erreur quadratique
print(model.summary())
```

Il s'agit d'un réseau dense à deux couches cachés.

La couche d'entrée a 100 neurones (dimension 2*50)
La première couche cachée a 300 neurones (dimension 2*150)
La deuxième couche cachée a 200 neurones (dimension 2*100)
La couche de sortie a 100 neurones (2*50)

Le "none" dans la dimension de sortie correspond au batch size. Il n'est pas défini et dépend du nombre de données que l'on passera simultanément dans le résseau. 

L'optimiser "adam" fonctionne très bien en général et ne necessite pas d'être changé

Plusieurs types de metrics sont disponibles suivant les applications. La documentation donne quelques informations à ce sujet : https://keras.io/api/metrics/ 

#Apprentissage


La séparation de la base de validation / apprentissage se fait ici. Un bon choix est de "validation_split=0.2" pour que 20% de la base d'apprentissage soit dédié à la validation

Le nombre d'epochs est réglé pour converger en évitant le sur apprentissage. Voir le résultat et l'augmentation du loss de validation pour détécter le surapprentissage 

```python
epochs=1000
history = model.fit(X, y, batch_size=32, epochs=epochs, verbose=1, validation_split=0.2)
```

Affichage de l'évolution de la fonction de coût et de la précision pendant l'apprentissage

```python
acc = history.history['mean_squared_error']
val_acc = history.history['val_mean_squared_error']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.xlabel("Number of iteration")
plt.ylabel("Value")
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.xlabel("Number of iteration")
plt.ylabel("Value")
plt.title('Training and Validation Loss')
plt.show()
```

On retrouve les mêmes courbes car la précision est définie par une erreur quadratique et le loss c'est l'erreur quadratique aussi. Ca dépend de notre "metric" et s'est normal pour une regression.

On visualise du surapprentissage car :

*   La base de données est trop petite
*   Le modèle est trop complexe (trop de paramètres)
*   Pas d'arret anticipé à la ré-augmentation du "loss" de validation (early stopping)

Si l'apprentissage n'a pas convergé, il faut augmenter le nombre d'epochs

Si l'erreur est encore trop élevé par rapport au cahier des charges, il faut modifier les hyperparametres afin d'améliorer l'apprentissage.

Il est possible de jouer sur :


*   Le type de réseau (dense, convolution, récurrent)
*   Le nombre de couches
*   Le nombre de neurones par couche
*   Les fonctions d'activation
*   La fonction de cout et les metrics
*   Le batch size, le learning rate
